# Code for Naive Bayes classifier

```python
def prior_prob(ytrain, label):
    return np.sum(ytrain==label)/ytrain.shape[0]
```

```python
prior_prob(ytrain,1)
```

```python
def cond_prob(x_train,y_train,feature_col, feature_val, label):
    
    xfilt = x_train[y_train==label]
    constraint= np.sum(xfilt[:,feature_col]==feature_val)
    prob_value = constraint/float(xfilt.shape[0])
    return prob_value
```

```python
def predict(x_train,y_train,x_test):
    preds = []
    classes = np.unique(ytrain)
    for ix in range(x_test.shape[0]):
        post_prob = []
    
        for jx in classes:
            likelihood = 1.0
#         print(x_test[ix])
            for kx in range(x_test[ix].shape[0]):
                cond = cond_prob(x_train,y_train,kx,x_test[ix,kx],jx)
                likelihood *= cond
            prior = prior_prob(y_train,jx)
            posterior_prob = likelihood*prior
#         print(posterior_prob)
            post_prob.append(posterior_prob)
        post_prob = np.array(post_prob)
#     print(post_prob)
        pred_label = post_prob.argmax()
        preds.append(pred_label)
    return preds
```

```python
predictions = predict(xtrain,ytrain,xtest)
```

```python
np.sum(predictions==ytest)/ytest.shape[0]
```
